# Data Analysis Template

Flow: imports -> load -> inspect -> clean -> analyze -> visualize -> summarize.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Update path before execution
df = pd.read_csv('data/raw/input.csv')
df.head()
```

```python
df.info()
df.describe(include='all').T
```

```python
# Example cleaning step
df = df.drop_duplicates().reset_index(drop=True)
df.isna().mean().sort_values(ascending=False).head(10)
```

```python
# Example analysis
summary = df.groupby(df.columns[0]).size().rename('count').sort_values(ascending=False)
summary.head(10)
```

```python
ax = summary.head(10).plot(kind='bar', figsize=(10, 4), title='Top categories')
ax.set_ylabel('count')
plt.tight_layout()
```
